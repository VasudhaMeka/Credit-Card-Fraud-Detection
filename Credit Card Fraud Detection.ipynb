```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
plt.figure(figsize = (25,25))
sns.heatmap(df.corr(),annot=True,fmt='.1f')
plt.show()
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.30,random_state=0)
```

```python
sc= StandardScaler()

x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)
```

### <i>Logistic Regression</i>

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
lr_acc_score = accuracy_score(y_test,y_pred_lr)
print(f"Accuracy score for Logistic Regression is {lr_acc_score}")
```

```python
print(classification_report(y_test,y_pred_lr))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, fmt='g');
```

### <i>Random Forest</i>

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
rf_acc_score = accuracy_score(y_test,y_pred_rf)
print(f'Accuracy score for Random Forest is {rf_acc_score}')
```

```python
print(classification_report(y_test,y_pred_rf))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred_rf),annot=True,fmt='g')
plt.show()
```
